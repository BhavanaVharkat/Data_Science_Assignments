# Excercise 1:Prime Number

```python
#Write a Python program that checks whether a given number is prime or not.
#A prime number is a natural number greater than 1 that has no 
#positive divisors other than 1 and itself.
```

```python
n=int(input("Enter the number:"))
for i in range(2,n):
    if n%i == 0:
        print(n,"is not prime")
        break
else:
    print(n,"is prime")
```

# Excercise 2:Product of Random Numbers

```python
#Develop a Python program that generates two random numbers 
#and asks the user to enter the product of these numbers. 
#The program should then check if the user's answer is correct and display an appropriate message.
```

```python
import random
n1 = random.randint(1,10)
n2 = random.randint(1,10)
print("First number:",n1)
print("Second number:",n2)
user_answer = int(input("Enter the product of two numbers:"))
if user_answer == n1*n2:
    print("Well done,correct answer!")
else:
    print("Wrong answer,try again:)")
```

# Exercise 3: Squares of Odd Numbers

```python
#Create a Python script that prints the squares of all even 
#or odd numbers within the range of 100 to 200.
#Choose either even or odd numbers and document your choice in the code.
```

```python
print("Squares of odd numbers from 200 to 300")
for i in range(200,301):
    if i % 2!=0:
        print(i,"=",i*i)
```

# Exercise 4:Hospital Billing System

```python
#A hospital charges its patients based on the following:
# •	Room charges per day: ₹2,000
# •	Doctor consultation fee (flat): ₹1,500
# •	Lab test charges: Based on the number of tests taken. Each test costs ₹300
# •	Medicine charges: Total cost of medicines provided
# •	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
# Write a Python program to calculate the final bill for a patient who:
# •	Stayed for 4 days
# •	Had 3 lab tests
# •	Had medicine charges of ₹2,400
```

```python
#Basic charges
room_charges = 2000
doctor_consultation_fees = 1500
lab_test_charges = 300

#Patient details
stayed_days = 4
lab_tests = 3
medicine_charges = 2400

#Billing
room_charges = room_charges * stayed_days
lab_charges = lab_test_charges * lab_tests
total_bill = room_charges + doctor_consultation_fees + lab_test_charges + medicine_charges

#Discount(If applicable)
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

#For output
print("Total bill before discount: ₹", total_bill)
print("Discount applied: ₹", discount)
print("Final bill to be paid: ₹", final_bill)
```

```python
# Given charges
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300

# Patient details
days_stayed = 4
number_of_tests = 3
medicine_charges = 2400

# Calculations
room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_cost * number_of_tests

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if applicable
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

# Output
print("Total bill before discount: ₹", total_bill)
print("Discount applied: ₹", discount)
print("Final bill to be paid: ₹", final_bill)
```
